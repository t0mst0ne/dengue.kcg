# Loading library and data

```r
library(R0)
dengue2015 <- read.csv("data/dengue2015_for_R.csv", head=TRUE,sep=",") 
dengue2015$t <- as.Date(dengue2015$t)
#check.incid(dengue2015$incid, dengue2015$t)
mGT <- generation.time("gamma", c(3,1.5))
dengue2015$incid
```

# Using Time-Dependent method 

```r

TD <- est.R0.TD(dengue2015$incid, mGT, begin=10, end=206, nsim=200)
plot(TD)

```

# Using various kind of methods for estimate R0

```r
est.R0.EG(dengue2015$incid, mGT, begin=1, end=200)
estR0<-estimate.R(dengue2015$incid, mGT, begin=1, end=200, methods=c("EG", "ML", "TD", "AR", "SB"),pop.size=100000, nsim=1000)
plot(estR0)
plotfit(estR0)

```
